```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
product_names=[]
product_prices=[]
product_ratings=[]
product_descr=[]
```

```python
page_no = input("Enter number of pages :")
for i in range(1, int(page_no)+1):
  url="https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles|Mobiles&requestId=0edc2bbe-d30a-45fc-80a0-a3a797f85ce0&as-searchtext=mobile&page="+str(i)
  req = requests.get(url)
  content = BeautifulSoup(req.content, "html.parser")
  name = content.find_all('div',{"class":"_4rR01T"})
  price = content.find_all('div',{"class":"_30jeq3 _1_WHN1"})
  rating = content.find_all('div',{"class":"_3LWZlK"})
  disc = content.find_all('div',{"class":"fMghEO"})
  print("Phones are in page : "+str(i))
  print(len(name), len(price), len(rating), len(disc))
  for i in name:
    product_names.append(i.text)
  for i in price:
    product_prices.append(i.text)
  for i in range(len(rating)):
    if i < len(name):
      product_ratings.append(rating[i].text)
  for i in disc:
    product_descr.append(i.text)
  print(len(name), len(price), len(product_ratings), len(disc))
```

```python
for i in product_names:  # product_prices, product_ratings, product_descr
      print(i)
```

```python
data = {'Product_Name':product_names, "Product_Price":product_prices, "Ratings": product_ratings, "Discription":product_descr}
df = pd.DataFrame(data)
```
